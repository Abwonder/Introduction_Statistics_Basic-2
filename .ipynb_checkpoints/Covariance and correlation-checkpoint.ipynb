```python
## channging to existing dataset checked up
df = sns.load_dataset("healthexp")
df.head()  ## it might take time sometimes as it loads from github repo
```

```python
### call needed library
import numpy as np
```

```python
np.cov(df)
```

```python
type(df)
```

```python
### this can be done this way, using numpy by calling out each column
np.cov([df["Year"], df["Spending_USD"], df["Life_Expectancy"]])
# Note: Just the column with the numerical variables, make it a 3d array
```

```python
### Best way to do it
df.cov()
# Observing the data only Spending_USD and Life_Expectancy, it shows that will
# increasing speniding their is increase in life_expectancy
```

```python
## Spearman Rank correlation 
df.corr(method = "spearman")
```

```python
# Peason correlation
df.corr()  ## default is pearson correlation
```

```python
# Peason correlation
df.corr(method = "pearson")
```
