### Question 5

Q5. How are measures of dispersion such as range, variance, and standard deviation used to describe
the spread of a dataset? Provide an example.

__Range__

This is used to express the highest and the lowest point of the values in a dataset which provides a rough estimate of the spread of the data. E.g between 20-50 the range is 50-20 = 30; 30 is the range

__Variance__

The variance is the average of the squared differences between each data point and the mean. It measures how much the data points are spread out from the mean. A high variance indicates that the data points are widely spread out, while a low variance indicates that the data points are clustered around the mean. For example, if the variance of a set of exam scores is 100, it indicates that the scores are widely dispersed around the mean score.

__Standard Deviation__

This measure is the square root of variance, which is the most used. A high standard deviation indicates that the data points are spread out, while a low standard deviation indicates that the data points are clustered around the mean. For example, if the standard deviation of a set of weights is 2 kg, it indicates that the weights are clustered around the mean weight with a relatively small spread.

Q6. What is a Venn diagram?

A Venn diagram is a graphical representation of the relationships between different sets of data. It consists of a series of overlapping circles or ellipses that represent the sets, with each circle representing a different set and the overlap representing the intersection of the sets. Venn diagrams are used to visualize the relationships between sets, especially the similarities and differences between them.

The area inside each circle or ellipse represents the elements or members of the set, and the overlapping region represents the elements that are common to both sets. The size of each circle or ellipse represents the relative size or proportion of the set.

Venn diagrams can be used for a variety of purposes, such as analyzing data, solving logic puzzles, and illustrating concepts in mathematics, statistics, and other fields. They are a useful tool for organizing information and identifying patterns and relationships between different sets of data.


### Question 7

Q7. For the two given sets A = (2,3,4,5,6,7) & B = (0,2,6,8,10). Find:
(i) A B
(ii) A ⋃ B

Answer

Explanation:


Given sets are:

A = {2, 3, 4, 5, 6, 7}
B = {0, 2, 6, 8, 10}

(i) A intersection B (A ⋂ B):

The intersection of two sets A and B is the set of all elements that are common to both sets. We can find the intersection of A and B by listing the elements that are present in both sets. In this case, we see that the only common element between A and B is 2. Therefore:

A ⋂ B = {2, 6}



(ii) A union B (A ⋃ B):

The union of two sets A and B is the set of all elements that are in either set A or set B, or both. We can find the union of A and B by combining all the elements from both sets and eliminating any duplicates. In this case, we have:

A ⋃ B = {0, 2, 3, 4, 5, 6, 7, 8, 10}

Therefore, the solution to the given problem is:

(i) A ⋂ B = {2}
(ii) A ⋃ B = {0, 2, 3, 4, 5, 6, 7, 8, 10}

```python
### Using Code 
## set of data

A = {2, 3, 4, 5, 6, 7}
B = {0, 2, 6, 8, 10}

# Find the intersection
A_intersect_B = A.intersection(B)
print("Intersection of A and B:", A_intersect_B)

# Find the union
A_union_B = A.union(B)
print("Union of A and B:", A_union_B)
```

### Question 8

Q8. What do you understand about skewness in data?

Skewness is the degree of asymmetry in a probability distribution is measured by a statistic called skewness. In other words, it quantifies the degree to which a dataset departs from the symmetric, skew-zero normal distribution.

When the mode (the most common value) is higher than the mean, we say that the distribution is skewed to the left. This means that the left-hand tail of the distribution is longer than the right-hand tail. The term "negative skewness" describes this phenomenon.


If the distribution is right-skewed, then the right-hand tail is longer and the mode is lower than the mean. Positive skewness is another term for this.

Pearson's skewness coefficient, Bowley's skewness coefficient, and the moment-based coefficient are all ways to quantify skewness. If the skewness coefficient is positive, the distribution is skewed to the right, and if it's negative, it's skewed to the left.

Due to its potential impact on the reliability and validity of statistical models and tests, skewness is an essential consideration in the field of statistics. For instance, the t-test may not work with a highly skewed dataset because of its dependence on normality assumptions.

### Question 10

Q10. Explain the difference between covariance and correlation. How are these measures used in
statistical analysis?

Covariance measures how two variables vary together. Specifically, it measures how much two variables move in relation to each other. If the two variables tend to increase or decrease together, their covariance will be positive, while if one variable tends to increase while the other decreases, their covariance will be negative. However, the size of the covariance does not tell us about the strength of the relationship between the two variables, as it depends on the scales of the variables.

While correlation, on the other hand, measures the strength and direction of the linear relationship between two variables, and it is normalized to a range of -1 to 1. A correlation coefficient of +1 indicates a perfect positive correlation, -1 indicates a perfect negative correlation, and 0 indicates no linear correlation. Correlation is a unitless measure and thus not affected by the scale of the variables.

Both covariance and correlation are used in statistical analysis to learn more about the connection between two numbers. Covariance can be used to measure how strongly two variables are related to one another, but its magnitude is highly unit-specific, making it hard to interpret. The strength and direction of the relationship between two variables can be quantified with correlation, making it the more popular of the two measures. Regression analysis, in which you try to predict one variable based on another variable or set of variables, makes heavy use of correlation.

Q11. What is the formula for calculating the sample mean? Provide an example calculation for a
dataset.

Answer

The sample mean is the average value of a dataset and is calculated by adding up all the values in the dataset and dividing by the number of observations in the dataset. The formula for calculating the sample mean is:

sample mean = (sum of all values in the dataset) / (number of observations in the dataset)

For example, suppose we have the following dataset of 5 values:

{2, 5, 6, 8, 10}

To calculate the sample mean, we would first add up all the values in the dataset:

2 + 5 + 6 + 8 + 10 = 31

Then we would divide the sum by the number of observations in the dataset, which in this case is 5:

sample mean = 31 / 5 = 6.2

Therefore, the sample mean for this dataset is 6.2.

### Question 13

Q13. How is covariance different from correlation?

Answer:

One important difference between covariance and correlation is that covariance is not standardized, meaning that its value depends on the scale of the variables. Correlation, on the other hand, is standardized, meaning that its value is always between -1 and 1, regardless of the scale of the variables. This makes correlation a more useful measure of the strength of the relationship between two variables.

Another important difference is that covariance can take on any value, while correlation is limited to the range between -1 and 1. This means that covariance can be difficult to interpret, while correlation is easier to interpret because it is standardized.

In summary, covariance and correlation are both measures of the relationship between two variables, but covariance measures the degree to which two variables move together, while correlation measures the strength and direction of the linear relationship between the two variables. Correlation is a more useful measure because it is standardized and easier to interpret.


### Question 14

Q14. How do outliers affect measures of central tendency and dispersion? Provide an example.

Answer

for central tendency
Outliers can significantly shift the mean, making it a less representative measure of the typical value in the dataset. The median, which is the middle value in the dataset, is less affected by outliers than the mean. However, if an outlier is very extreme, it can still have an impact on the median.


Measures of dispersion, such as the range, variance, and standard deviation, are also affected by outliers. The range, which is the difference between the maximum and minimum values in the dataset, can be significantly skewed by an outlier. The variance and standard deviation, which measure how much the data points deviate from the mean, can also be impacted by outliers. If an outlier is very extreme, it can increase the variance and standard deviation, making the data appear more spread out than it actually is.


__Example__


$30,000, $35,000, $40,000, $45,000, $50,000, $55,000, $60,000, $65,000, $70,000, $75,000, $80,000, $85,000, $90,000, $95,000, $100,000

The mean salary in this dataset is $62,000. However, if there was an outlier of $1,000,000, the mean salary would be significantly skewed to $70,000, making it a less representative measure of the typical salary in the company. Additionally, the variance and standard deviation of the dataset would be increased due to the outlier, making the data appear more spread out than it actually is.
